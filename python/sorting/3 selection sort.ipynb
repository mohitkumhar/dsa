```python
import random
import timeit

def find_ind_min(nums, start):
    minimum = nums[start]
    index = start
    for i in range(start, len(nums)):
        if nums[i] < minimum:
            minimum = nums[i]
            index = i
    return minimum, index

def selection_sort(nums):
    n = len(nums)
    for i in range(n):
        minimum, index = find_ind_min(nums, i)
        nums[i], nums[index] = nums[index], nums[i]
    return nums

# Generate a list of 1000 random numbers
a = [random.randint(0, 1000) for _ in range(1000)]

# Measure the running time using timeit
running_time = timeit.timeit('selection_sort(a.copy())', globals=globals(), number=1)

# Print the sorted list and the running time
sorted_a = selection_sort(a.copy())
print(sorted_a)
print(f"Running time: {running_time:.6f} seconds")
```
